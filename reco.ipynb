```python
# Importation
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
data = pd.read_csv('./dataset.csv')
x_train = data.drop(['Valeur'], axis=1)[:80000]
y_train = data['Valeur'][:80000]
x_test = data.drop(['Valeur'], axis=1)[-20000:]
y_test = data['Valeur'][-20000:]
(x_train, y_train), (x_test, y_test)
```

```python
# Create a model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=4),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1)
])

model
```

```python
model.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adam(learning_rate=0.1))
```

```python
results = model.fit(x_train, y_train, epochs=5, verbose=False)
print("-------------- Trained Finised ----------------")
```

```python
import matplotlib.pyplot as plt
plt.xlabel("Epoch Number")
plt.ylabel("Loss Magnitude")
plt.plot(results.history["loss"])
```

```python
print(model.predict(np.array([5, 0, 1, 5]).reshape(-1, 4)))
```
